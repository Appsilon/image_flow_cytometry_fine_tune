```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
channel_name = {
    "Ch01":"BF",
"Ch02":"Ab",
#"Ch03":"CD45",
"Ch04":"F-Actin",
"Ch06":"MHCII",
"Ch07":"CD3",
"Ch11":"P-CD3zeta",
"Ch12":"Live/Dead"

}

markers = channel_name.values()
```

```python
exp1 = pd.read_excel("exp1.xlsx", index_col=0)
exp1 = exp1.rename(channel_name, axis=0)
exp1 = exp1.rename(channel_name, axis=1)
exp1 = exp1.loc[markers,markers]
```

```python
exp2 = pd.read_excel("exp2.xlsx", index_col=0)
exp2 = exp2.rename(channel_name, axis=0)
exp2 = exp2.rename(channel_name, axis=1)
exp2 = exp2.loc[markers,markers]
```

```python
exp3 = pd.read_excel("exp3.xlsx", index_col=0)
exp3 = exp3.rename(channel_name, axis=0)
exp3 = exp3.rename(channel_name, axis=1)
exp3 = exp3.loc[markers,markers]
```

```python
exp4 = pd.read_excel("exp4.xlsx", index_col=0)
exp4 = exp4.rename(channel_name, axis=0)
exp4 = exp4.rename(channel_name, axis=1)
exp4 = exp4.loc[markers,markers]
```

```python
fig, ax = plt.subplots(nrows = 2, ncols = 2, figsize=(10,10), sharex=True, sharey=True)

ax[0,0] = sns.heatmap(exp1, annot=True, cbar=False, ax=ax[0,0], cmap="Greys", fmt=".3f")
ax[0,0].set_title("Experiment I")

ax[0,1] = sns.heatmap(exp2, annot=True, cbar=False, ax=ax[0,1], cmap="Greys", fmt=".3f")
ax[0,1].set_title("Experiment II")

ax[1,0] = sns.heatmap(exp3, annot=True, cbar=False, ax=ax[1,0], cmap="Greys", fmt=".3f")
ax[1,0].set_title("Experiment III")

ax[1,1] = sns.heatmap(exp4, annot=True, cbar=False, ax=ax[1,1], cmap="Greys", fmt=".3f")
ax[1,1].set_title("Experiment IV")

plt.savefig('confusion matrix.svg', bbox_inches="tight")
plt.savefig('confusion matrix.png', bbox_inches="tight")
```
